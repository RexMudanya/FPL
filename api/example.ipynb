```python
d = pd.DataFrame(all_data['history'])
plt.plot(d['total_points'], label='points')
plt.plot(d['bonus'], label='bonus')
plt.legend()
plt.title('points + bonus per GW')
plt.ylabel('points')
plt.xlabel('gw')
plt.show()
```

```python
plt.plot(d['value'])
plt.title('value per GW')
plt.ylabel('value')
plt.xlabel('gw')
plt.show()
```

```python
plt.plot(d['transfers_in'], label='in')
plt.plot(d['transfers_out'], label='out')
plt.title('transfers in vs out')
plt.legend()
plt.ylabel('transfers')
plt.xlabel('gw')
plt.show()
```

```python
plt.plot(d['goals_scored'], label='goals')
plt.plot(d['assists'], label='assists')
plt.title('goals and assists')
plt.legend()
plt.ylabel('goals + assists')
plt.xlabel('gw')
plt.show()
```
